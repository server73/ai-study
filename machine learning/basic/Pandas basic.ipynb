```python
train["Familysize"] = train["SibSp"] + train["Parch"] + 1
train[["SibSp", "Parch", "Familysize"]].head()
```

```python
train.head()
```

```python
train["Child"] = train["Age"] < 18
```

```python
train.head()
```

```python
train["Nationality_FR"] = train["Embarked"] == 'C'
train["Nationality_UK"] = train["Embarked"].isin(['Q','S'])
train.head()
```

```python
train.loc[train["Embarked"] == "C", "Nationality"] = "France"
train.loc[train["Embarked"].isin(["Q", "S"]), "Nationality"] = "England"
train[["Embarked", "Nationality"]].head()
```

```python
train["Fare_Low"] = train["Fare"] < 15
train["Fare_Medium"] = (train["Fare"] >= 15) & (train["Fare"] < 35)
train["Fare_High"] = train["Fare"] >= 35
train[["Fare", "Fare_Low", "Fare_Medium", "Fare_High"]].head()
```

```python
mean_age = train["Age"].mean()
train.loc[train["Age"].isnull(), "Age"] = mean_age
train.head()
```
