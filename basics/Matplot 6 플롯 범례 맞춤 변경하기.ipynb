```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
```

```python
x = np.linspace(0,10,100)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
plt.legend();
```
